## Synch Time

```python
def plot_synch_avg_comparison_general(eth_data: dict, opx_data: dict, model_name='Base'):
    """
    Plots a grouped bar chart of synch_avg_time (excluding 0) for various worker counts,
    comparing Ethernet vs OPX for a given ViT model.

    Parameters:
    - eth_data: Dict[int, pd.DataFrame], mapping worker count to Ethernet DataFrame
    - opx_data: Dict[int, pd.DataFrame], mapping worker count to OPX DataFrame
    - model_name: 'Base', 'Large', or 'Huge'
    """

    def avg_synch(df):
        return df.loc[df['synch_avg_time'] != 0, 'synch_avg_time'].mean()

    # Sorted list of common worker counts
    worker_counts = sorted(set(eth_data.keys()).intersection(opx_data.keys()))

    ethernet_vals = [avg_synch(eth_data[w]) for w in worker_counts]
    opx_vals = [avg_synch(opx_data[w]) for w in worker_counts]
    labels = [f'{w} Workers' for w in worker_counts]
    x = np.arange(len(worker_counts))
    bar_width = 0.35

    # Plot
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(x - bar_width/2, ethernet_vals, width=bar_width, label='Ethernet', color='#1f77b4')
    ax.bar(x + bar_width/2, opx_vals, width=bar_width, label='OPX', color='#d62728')

    # Annotate bars
    for i in range(len(x)):
        ax.text(x[i] - bar_width/2, ethernet_vals[i], f"{ethernet_vals[i]:.2f}", ha='center', va='bottom')
        ax.text(x[i] + bar_width/2, opx_vals[i], f"{opx_vals[i]:.2f}", ha='center', va='bottom')

    # Style
    ax.set_title(f'ViT-{model_name} - Parameters Averaging Time')
    ax.set_ylabel('Time (s)')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```
