```python
def plot_synch_avg_comparison(df_2_eth1, df_4_eth1, df_2_ib0, df_4_ib0, model_name='Base'):
    """
    Plots a grouped bar chart of synch_avg_time (excluding 0) for 2 and 4 workers,
    comparing Ethernet vs OPX for a given ViT model.

    Parameters:
    - df_2_eth1: DataFrame for 2 workers, Ethernet
    - df_4_eth1: DataFrame for 4 workers, Ethernet
    - df_2_ib0: DataFrame for 2 workers, OPX
    - df_4_ib0: DataFrame for 4 workers, OPX
    - model_name: 'Base', 'Large', or 'Huge'
    """

    # Helper to compute average excluding zeros
    def avg_synch(df):
        return df.loc[df['synch_avg_time'] != 0, 'synch_avg_time'].mean()

    # Compute means
    ethernet_vals = [avg_synch(df_2_eth1), avg_synch(df_4_eth1)]
    opx_vals = [avg_synch(df_2_ib0), avg_synch(df_4_ib0)]
    labels = ['2 Workers', '4 Workers']
    x = np.arange(len(labels))
    bar_width = 0.35

    # Plot
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.bar(x - bar_width/2, ethernet_vals, width=bar_width, label='Ethernet', color='#1f77b4')
    ax.bar(x + bar_width/2, opx_vals, width=bar_width, label='OPX', color='#d62728')

    # Annotate bars
    for i in range(len(x)):
        ax.text(x[i] - bar_width/2, ethernet_vals[i], f"{ethernet_vals[i]:.2f}", ha='center', va='bottom')
        ax.text(x[i] + bar_width/2, opx_vals[i], f"{opx_vals[i]:.2f}", ha='center', va='bottom')

    # Style
    ax.set_title(f'ViT-{model_name} - Parameters Averaging Time')
    ax.set_ylabel('Time (s)')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# ViT-Base
plot_synch_avg_comparison(
    ddp_2_b_64_8_eth1, ddp_4_b_64_8_eth1,
    ddp_2_b_64_8_ib0, ddp_4_b_64_8_ib0,
    model_name='Base'
)

# ViT-Large
plot_synch_avg_comparison(
    ddp_2_l_64_4_eth1, ddp_4_l_64_4_eth1,
    ddp_2_l_64_4_ib0, ddp_4_l_64_4_ib0,
    model_name='Large'
)

# ViT-Huge
plot_synch_avg_comparison(
    ddp_2_h_64_8_eth1, ddp_4_h_64_8_eth1,
    ddp_2_h_64_8_ib0, ddp_4_h_64_8_ib0,
    model_name='Huge'
)
```
